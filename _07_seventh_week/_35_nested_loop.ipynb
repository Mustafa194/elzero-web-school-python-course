### <mark style="background-color: #CD6354; color: white; padding: 0 1rem;">i for numbering, j for skill name, people[person][skill] for the percentage</mark>

```python
for person in people:
    print(f"{person.capitalize()} Skills Are:")
    i = 1
    for skill in people[person]:
        print(f"{i} {skill.upper() if len(skill) < 5 else skill} {people[person][skill]}")
        i += 1
    print()
```

### <mark style="background-color: #CD6354; color: white; padding: 0 1rem;">Example</mark>

```python
people = {
	'Mustafa': {
		'First Name':'Mustafa',
		'Middle Name':'Ahmed',
		'Last Name':'Mohammed',
		'Skills': {
			'Html': '60%',
			'Css': '40%',
			'JS': '40%',
			'C++': '60%',
			'Python': '30%',
			'MSSQL': '20%'
		}
	},
	'Ahmed': {
		'First Name':'Mustafa',
		'Middle Name':'Ahmed',
		'Last Name':'Mohammed',
		'Skills': {
			'Adobe Photoshop': '55%',
			'Adobe Illustrator': '40%',
			'Adobe XD': '65%',
			'Adobe Premiere': '80%',
			'Adobe After Effects': '70%'
		}
	},
	'Mahmood': {
		'First Name':'Mustafa',
		'Middle Name':'Ahmed',
		'Last Name':'Mohammed',
		'Skills': {
			'Html': '90%',
			'Css': '85%',
			'JS': '85%',
			'Sass': '70%',
			'React.js': '75%',
			'Node.js': '70%',
			'Express.js': '60%',
			'MongodDB': '60%',
			'MySQL': '80%',
			'Command Line': '80%'
		}
	},
	'Ameer': {
		'First Name':'Mustafa',
		'Middle Name':'Ahmed',
		'Last Name':'Mohammed',
		'Skills': {
			'Networking': '90%',
			'Windows 10 Troubleshooting': '90%',
			'Command Line': '90%',
			'Linux': '75%',
			'CyberSecurity': '80%',
			'Ethical Hacking': '80%',
			'Artificial Intelligence': '80%',
			'Python': '90%',
			'JS': '90%',
			'MySQL': '90%',
			'MSSQL': '90%',
			'MariaDB': '90%',
			'PostgreSQL': '90%'
		}
	}
}
```

```python
for person in people:
    print(f"{person} Properties are:")
    for prop in people[person]:
        if type(people[person][prop]) == dict:
            print(f'Skills:')
            for skill in people[person][prop]:
                print(f"\t{skill}: {people[person][prop][skill]}".expandtabs(4))
        else:
            print(f"{prop}: {people[person][prop]}")
    print("_" * 60)
    print() # prints a new line
```
